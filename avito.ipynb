```python
train_df = train_df.dropna()
```

```python
preprocessed_path = 'data/train.preprocessed.csv'
if not os.path.exists(preprocessed_path):
    for col in train_df.columns:
        if train_df[col].dtype.kind == 'O':
            train_df[col] = train_df[col].str.replace('\n', ' ')

    train_df.to_csv(preprocessed_path, index=False)
```

```python
def train_input_fn():
    return tf.contrib.data.make_csv_dataset('data/train.preprocessed.csv',
                                            batch_size=10,
                                            num_epochs=1,
                                            shuffle=True,
                                            label_name='deal_probability')

model = tf.estimator.DNNRegressor(feature_columns=[tf.feature_column.numeric_column('price')],
                                  hidden_units=[])
model.train(train_input_fn)
```
